```python

from google.colab import files
uploaded = files.upload(
)
import pandas as pd
file = pd.read_csv('fandango_scrape.csv')
file.head(3)
file.describe()
number_5star = file[file["STARS"]== 5.0]
print(len(number_5star))
file["STARS"].value_counts()
a_mean = file[file["VOTES"] < file["VOTES"].mean()]
a_mean["STARS"].value_counts()
a_5k = file[file["VOTES"] < 5000.0]
a_5k["STARS"].value_counts()
a_1k = file[file["VOTES"] < 1000.0]
a_1k["STARS"].value_counts()
a_500 = file[file["VOTES"] < 500.0]
a_500["STARS"].value_counts()
a_50 = file[file["VOTES"] < 50.0]
a_50["STARS"].value_counts()
a_10 = file[file["VOTES"] < 10.0]
a_10["STARS"].value_counts()
a_5 = file[file["VOTES"] < 5.0]
a_5["STARS"].value_counts()
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(14, 18))

# Mean votes
plt.subplot(4, 2, 1)
a_mean["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < Mean")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 5000
plt.subplot(4, 2, 2)
a_5k["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 5000")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 1000
plt.subplot(4, 2, 3)
a_1k["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 1000")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 500
plt.subplot(4, 2, 4)
a_500["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 500")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 50
plt.subplot(4, 2, 5)
a_50["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 50")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 10
plt.subplot(4, 2, 6)
a_10["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 10")
plt.xlabel("Stars")
plt.ylabel("Count")

# < 5
plt.subplot(4, 2, 7)
a_5["STARS"].value_counts().sort_index().plot(kind="bar")
plt.title("Votes < 5")
plt.xlabel("Stars")
plt.ylabel("Count")

plt.tight_layout()
plt.show()


from google.colab import files
uploaded = files.upload(
)

file2 =  pd.read_csv('movie_ratings_16_17.csv')
after = file2[['movie', 'year', 'fandango']].copy()

after.head(3)

after["fandango"].value_counts()

# Create figure
plt.figure(figsize=(14, 18))

# Mean votes
plt.subplot(4, 2, 1)
after["fandango"].value_counts().sort_index().plot(kind="bar")
plt.title("stars vs count")
plt.xlabel("Stars")
plt.ylabel("Count")
```
